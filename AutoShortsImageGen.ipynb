```python

from diffusers import DiffusionPipeline, FluxPipeline
import torch

```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

model = "F"

if model == "FLUX":
  diffusionPipe = FluxPipeline.from_pretrained("black-forest-labs/FLUX.1-schnell", torch_dtype=torch.bfloat16)
# LOAD DIFFUSION MODEL
else:
  model_id = "playgroundai/playground-v2.5-1024px-aesthetic"
  diffusionPipe = DiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16,variant="fp16").to("cuda")
```

```python

import configparser
import os

def create_config(base_path):
    config = configparser.ConfigParser()
    # Add sections and key-value pairs
    config['General'] = {'curr_vid': 1}
    # Write the configuration to a file
    with open(base_path+'/images_config.ini', 'w') as configfile:
        config.write(configfile)

base_path = '/content/drive/MyDrive/AI'
#create_config(base_path)

def set_video(base_path,curr_vid):
  config = configparser.ConfigParser()
  config.read(base_path+'/images_config.ini')
  config['General']['curr_vid'] = str(curr_vid)
  with open(base_path+'/images_config.ini', 'w') as configfile:
    config.write(configfile)

def get_video(base_path):
  config = configparser.ConfigParser()
  config.read(base_path+'/images_config.ini')
  curr_vid = config['General']['curr_vid']
  return curr_vid
def get_prompts_video(base_path):
  config = configparser.ConfigParser()
  config.read(base_path+'/prompt_config.ini')
  curr_vid = config['General']['curr_vid']
  return curr_vid

curr_vid = get_video(base_path)

curr_vid_path = os.path.join(base_path,curr_vid)
if not os.path.exists(f"{curr_vid_path}/images"):
  os.mkdir(f"{curr_vid_path}/images")
```

```python
if get_video(base_path) <  get_prompts_video(base_path):
  set_video(base_path,int(curr_vid)+1)
  curr_vid = get_video(base_path)
  curr_vid_path = os.path.join(base_path,curr_vid)
  os.mkdir(f"{curr_vid_path}/images")
```

```python
Auto_gen = True
if Auto_gen:
  while int(get_video(base_path)) <  int(get_prompts_video(base_path)):
    set_video(base_path,int(curr_vid)+1)
    curr_vid = get_video(base_path)
    curr_vid_path = os.path.join(base_path,curr_vid)
    if not os.path.exists(f"{curr_vid_path}/images"):
      os.mkdir(f"{curr_vid_path}/images")

    with open(f"{curr_vid_path}/img_prompts.txt", "r") as f:
        responses = f.readlines()
    with open(f"{curr_vid_path}/prompts.txt", "r") as f:
        prompts = f.readlines()

    for i,response in enumerate(responses):
        if i>len(prompts):
            break
        response_img = diffusionPipe(prompt=response,num_inference_steps=20).images[0]

        response_img.save(f"{curr_vid_path}/images/{i+1}.png")
else:
  which_video = 5
  vid_path = os.path.join(base_path,str(which_video))
  with open(f"{vid_path}/img_prompts.txt", "r") as f:
    responses = f.readlines()
  with open(f"{vid_path}/prompts.txt", "r") as f:
      prompts = f.readlines()

  for i,response in enumerate(responses):
      if i>len(prompts):
          break
      response_img = diffusionPipe(prompt=response,num_inference_steps=20).images[0]

      response_img.save(f"{vid_path}/images/{i+1}.png")
```
